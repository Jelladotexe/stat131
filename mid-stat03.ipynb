3.1 บริษัทให้บริการด้าน IT ครบวงจรแห่งหนึ่งได้เก็บรวบรวมข้อมูลค่าซ่อมบำรุงอุปกรณ์คอมพิวเตอร์แม่ข่ายและจำนวนลูกค้าของบริษัทในปีที่ผ่านมาเป็นดังไฟล์ asm3_1_itcost.csv  
จงแสดงวิธีหาค่ามัธยฐานและฐานนิยมของค่าซ่อมบำรุงอุปกรณ์คอมพิวเตอร์แม่ข่าย

```python
df_it = pd.read_csv('https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/asm3_1_itcost.csv')
df_it.head()
```

```python
# แก้ข้อมูล
df_it['maintenance_cost'] = df_it['maintenance_cost'].replace('<=5000','1-5000')
df_it['maintenance_cost'] = df_it['maintenance_cost'].replace('>=30001','30001-35000')
df_it[['main_from','main_to']] = df_it.maintenance_cost.str.split('-',expand=True).astype(int)
df_it
```

```python
# FINDING MED
# สร้างความถี่สะสม
df_it['num_cus_cum'] = df_it['num_customer'].cumsum()

# หาตน. med
med_pos = df_it['num_customer'].sum() / 2

# หาตนชั้น med
for i in df_it.index:
  if df_it.num_cus_cum[i] >= med_pos:
    e=i
    break

lower = (df_it.main_from[e]+df_it.main_to[e-1])/2
med = lower+((med_pos - df_it.num_cus_cum[e-1]) / df_it.num_customer[e])*5000

print("Med =", med)
```

```python
# FINDING MODE
e = df_it['num_customer'].idxmax()
lower = (df_it.main_from[e]+df_it.main_to[e-1])/2
de_minus = np.abs(df_it.num_customer[e] - df_it.num_customer[e-1])
de_plus = np.abs(df_it.num_customer[e] - df_it.num_customer[e+1])
mode = lower + (de_minus / (de_minus + de_plus)) * 5000
print("Mode =", mode)
```

3.2 สุ่มตัวอย่างนักศึกษามาจำนวนหนึ่ง พบว่าคะแนนสอบวิชาสถิติ เป็นดังตาราง 'asm3_2_score.csv' จงหาค่า

**3.2.1 ค่าเฉลี่ย**

```python
df_score = pd.read_csv('https://raw.githubusercontent.com/ThammakornS/ProgStat/main/dataset/asm3_2_score.csv')
df_score.head()
```

```python
boundary = df_score.score.str.split('-',expand=True).astype(int)
df_score['x'] = (boundary[0] + boundary[1])/2
df_score['fx'] = df_score['num_student'] * df_score['x']
mean = df_score['fx'].sum() / df_score['num_student'].sum()
print("ค่าเฉลี่ย =",mean)
#df_score
```

**3.2.2 ค่ามัธยฐาน**

```python
#your code
df_score[['score_from','score_to']] = df_score.score.str.split('-',expand=True).astype(int)
df_score['student_cumulate'] = df_score['num_student'].cumsum()
med_pos = df_score['num_student'].sum()/2

for i in df_score.index:
  if df_score.student_cumulate[i] >= med_pos:
    e = i
    break
lower = (df_score.score_from[e] + df_score.score_to[e-1])/2
med = lower + ((med_pos - df_score.student_cumulate[e-1]) / df_score.num_student[e]) * 5
print("ค่ามัธยฐาน =", med)
#df_score
```

**3.2.3 ค่าฐานนิยม**

```python
e = df_score['num_student'].idxmax()
lower = (df_score.score_from[e] + df_score.score_to[e-1])/2
de_minus = np.abs(df_score['num_student'][e] - df_score['num_student'][e-1])
de_plus = np.abs(df_score['num_student'][e] - df_score['num_student'][e+1])
mode = lower+ (de_minus / (de_minus+de_plus)) * 5
print("ค่าฐานนิยม =", mode)
```

**3.2.4 Q<sub>1</sub> และ Q<sub>3</sub>**

```python
n = df_score['num_student'].sum()
q1_pos = n*1 /4
q3_pos = n*3/4

e1 = 1
e2 = 4
lower1 = (df_score.score_to[e1-1] + df_score.score_from[e1]) / 2
lower2 = (df_score.score_to[e2-1] + df_score.score_from[e2]) / 2

q1 = lower1 + ((q1_pos - df_score.student_cumulate[e1-1]) / df_score.num_student[e1]) * 5
q3 = lower2 + ((q3_pos - df_score.student_cumulate[e2-1]) / df_score.num_student[e2]) * 5

print("Q1 =", q1)
print("Q3 =", q3)
```

**3.2.5 D<sub>7</sub>**

```python
d7_pos = n*7/10
e = 4
lower = (df_score.score_to[e-1] + df_score.score_from[e]) / 2
d7 = lower + ((d7_pos - df_score.student_cumulate[e-1]) / df_score.num_student[e]) * 5
print("D7 =",d7)
```

**3.2.6 P<sub>75</sub>**

```python
#your code
p75_pos = n*75 / 100
e = 4
lower = (df_score.score_to[e-1] + df_score.score_from[e]) / 2
p75 = lower + ((p75_pos - df_score.student_cumulate[e-1]) / df_score.num_student[e]) * 5
print("P75 =",p75)
df_score
```

**3.2.7 พิสัย**

```python
#your code
up = df_score['score_from'].min() - 0.5
down = df_score['score_to'].max() + 0.5
range = down - up
print("พิสัย =", range)
```

**3.2.8 ส่วนเบี่ยงเบนควอร์ไทล์**

```python
QD = (q3-q1)/2
print("ส่วนเบี่ยงเบนควอร์ไทล์ =",QD)
```

**3.2.9 ส่วนเบี่ยงเบนมาตราฐานและความแปรปรวน**

```python
#your code
df_score['x^2'] = df_score['x']**2
df_score['fx^2'] = df_score['num_student'] * df_score['x^2']

fx2 = df_score['fx^2'].sum()
nx2 = df_score['num_student'].sum()* (mean)**2

s = (fx2-nx2) / (df_score['num_student'].sum() - 1)
print(f"ส่วนเบี่ยงเบนมาตราฐาน = {np.sqrt(s)} ความแปรปรวน = {s}")
df_score
```

**3.2.10 สัมประสิทธิ์ของการแปรผัน**

```python
cv = np.sqrt(s) / mean * 100
print("สัมประสิทธิ์ของการแปรผัน =", cv)
```
